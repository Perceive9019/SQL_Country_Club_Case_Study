```python
# Q3: Write an SQL query to show a list of facilities that charge a fee to members,
# where the fee is less than 20% of the facility's monthly maintenance cost.
# Return the facid, facility name, member cost, and monthly maintenance of the
# facilities in question.

query3 = """
SELECT facid, name, membercost, monthlymaintenance, (membercost/monthlymaintenance)*100 AS per_month_cost 
FROM Facilities 
WHERE membercost>0.0 AND (membercost/monthlymaintenance)*100<20;
"""
q3resultset = pd.read_sql_query(query3, engine, index_col='facid')
q3resultset
```

```python
# Q5: Produce a list of facilities, with each labelled as
# 'cheap' or 'expensive', depending on if their monthly maintenance cost is
# more than $100. Return the name and monthly maintenance of the facilities
# in question.

query5 = """
SELECT name, monthlymaintenance, CASE WHEN (monthlymaintenance > 100) THEN "expensive" ELSE "cheap" END AS label 
FROM Facilities;
"""

q5resultset = pd.read_sql_query(query5, engine)
q5resultset
```

```python
# Q7: Produce a list of all members who have used a tennis court.
# Include in your output the name of the court, and the name of the member
# formatted as a single column. Ensure no duplicate data, and order by
# the member name.

query7 = """
SELECT DISTINCT fac.name AS facility, mem.firstname || " " || mem.surname AS member 
FROM Members AS mem 
JOIN Bookings AS book 
ON mem.memid=book.memid 
JOIN Facilities AS fac 
ON book.facid = fac.facid 
WHERE book.facid IN (0, 1) 
ORDER BY member;
"""

q7resultset = pd.read_sql_query(query7, engine)
q7resultset
```

```python
# Q8: Produce a list of bookings on the day of 2012-09-14 which
# will cost the member (or guest) more than $30. Remember that guests have
# different costs to members (the listed costs are per half-hour 'slot'), and
# the guest user's ID is always 0. Include in your output the name of the
# facility, the name of the member formatted as a single column, and the cost.
# Order by descending cost, and do not use any subqueries.

query8 = """
SELECT fac.name AS facility_name, fac.membercost, fac.guestcost, book.slots, book.starttime, 
CASE WHEN mem.surname = "GUEST" THEN mem.surname ELSE mem.surname || " " || mem.firstname end AS member_name, 
CASE WHEN mem.memid=0 THEN fac.guestcost*book.slots ELSE fac.membercost*book.slots END AS cost_per_day 
FROM Bookings AS book 
JOIN Facilities AS fac 
ON fac.facid = book.facid 
JOIN Members AS mem 
ON mem.memid = book.memid 
WHERE (book.starttime LIKE "2012-09-14%" 
AND mem.memid = 0 
AND fac.guestcost*book.slots > 30) 
ORDER BY cost_per_day DESC;
"""

q8resultset = pd.read_sql_query(query8, engine)
q8resultset
```

```python
# Q9: This time, produce the same result as in Q8, but using a subquery.

query9 = """
SELECT subquery.facility_name, subquery.member_name, subquery.cost_per_day 
FROM (SELECT book.starttime, mem.firstname || " " || mem.surname AS member_name, fac.name AS facility_name, 
CASE WHEN mem.memid=0 THEN fac.guestcost*book.slots ELSE fac.membercost END AS cost_per_day 
FROM Bookings AS book 
JOIN Facilities AS fac 
ON fac.facid = book.facid 
JOIN Members AS mem 
ON mem.memid = book.memid) subquery 
WHERE subquery.starttime LIKE "2012-09-14%" 
AND subquery.cost_per_day > 30 
ORDER BY cost_per_day DESC;
"""

q9resultset = pd.read_sql_query(query9, engine)
q9resultset
```

```python
# Q10: Produce a list of facilities with a total revenue less than 1000.
# The output of facility name and total revenue, sorted by revenue. Remember
# that there's a different cost for guests and members! 

query10 = """
SELECT subquery.facility_name, subquery.total_revenue 
FROM (SELECT fac.name AS facility_name, SUM(CASE WHEN mem.memid = 0 THEN fac.guestcost*slots ELSE fac.membercost end) AS total_revenue 
FROM Bookings AS book 
JOIN Facilities AS fac 
ON fac.facid = book.facid 
JOIN Members AS mem 
ON mem.memid = book.memid 
GROUP BY facility_name) subquery 
WHERE total_revenue < 1000 
ORDER BY total_revenue DESC;
"""

q10resultset = pd.read_sql_query(query10, engine)
q10resultset
```

```python
# Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order.

query11 = """
SELECT a.surname || ' ' || a.firstname AS members, b.surname || ' ' || b.firstname AS recommended_by 
FROM Members a, Members b 
WHERE a.recommendedby > 0 AND a.recommendedby = b.memid 
ORDER BY b.surname;
"""

q11resultset = pd.read_sql_query(query11, engine)
q11resultset
```

```python
# Q12: Find the facilities with their usage by member, but not guests 

query12 = """
SELECT name, firstname || ' ' || surname AS member_name, COUNT(surname) AS usage 
FROM Bookings 
LEFT JOIN Facilities 
USING(facid) 
LEFT JOIN Members 
USING(memid) 
WHERE memid != 0 
GROUP BY name, member_name;
"""

q12resultset = pd.read_sql_query(query12, engine)
q12resultset
```

```python
# Q13: Find the facilities usage by month, but not guests
query13 = """
        SELECT book.facid, fac.name, strftime('%m', book.starttime) as month, SUM(book.slots) AS usage_by_month
        FROM Bookings AS book
        INNER JOIN Facilities AS fac 
        ON book.facid = fac.facid
        WHERE book.memid !=0
        GROUP BY book.facid, month;
        """

q13resultset = pd.read_sql_query(query13, engine)
```
